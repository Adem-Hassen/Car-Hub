```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def add_cars(articles,df):
    for article in articles:
        car_name=article.find("div",class_="inner").find("div",class_="thumb-caption").find("h2")
        price=article.find("div",class_="inner").find("div",class_="item-foot").find("div",class_="price")
        mileage=article.find("div",class_="inner").find("div",class_="thumb-caption").find("ul").find("li",class_="road")
        year=article.find("div",class_="inner").find("div",class_="thumb-caption").find("ul").find("li",class_="year")
        gearbox=article.find("div",class_="inner").find("div",class_="thumb-caption").find("ul").find("li",class_="boite")
        transmission=article.find("div",class_="inner").find("div",class_="thumb-caption").find("ul").find("li",class_="transmission")
        tax_horsepower=article.find("div",class_="inner").find("div",class_="thumb-caption").find("ul").find("li",class_="horsepower")
        energy=article.find("div",class_="inner").find("div",class_="thumb-caption").find("ul").find("li",class_="fuel")
        governorate=article.find("div",class_="inner").find("div",class_="thumb-infos").find("span")
        description=article.find("div",class_="inner").find("div",class_="description").find("div").find("p")
        image=article.find("div",class_="carousel-wrapper").find("img",class_="thumb").get("src")
        link="https://www.automobile.tn"+article.find("a",class_="occasion-link-overlay").get("href")
        
        row={"Model":car_name.text.strip(),"Price(DT)":price.text.strip().replace("DT",""),"Mileage(Km)":mileage.text.strip(),"Year":year.text.strip(),"Gearbox":gearbox.text.strip(),"Transmission":transmission.text.strip(),"Tax Horsepower(CV)":tax_horsepower.text.strip(),"Energy":energy.text.strip(),"Governorate":governorate.text.strip(),"Description":description.text.strip(),"Image":image.strip(),"Original Post":link}
        df.loc[len(df)]=row
```

```python
cars_df=pd.DataFrame(columns=["Model","Price(DT)","Mileage(Km)","Year","Gearbox","Transmission","Tax Horsepower(CV)","Energy","Governorate","Description","Image","Original Post"])
size=500
for i in range(1,size):
    if i%10==0:
        print(f'{(i/size)*10}%')
    page=requests.get(f"https://www.automobile.tn/fr/occasion/s=/{i}")
    scraped=BeautifulSoup(page.text,"html.parser")
    articles_container=scraped.find("div",class_="articles")
    articles=articles_container.find_all("div",class_="occasion-item-v2")
    add_cars(articles,cars_df)
    
```

```python
cars_df=cars_df.drop_duplicates()
cars_df["Price(DT)"] = cars_df["Price(DT)"].str.replace(r'[^\d.]', '', regex=True).astype(float)
cars_df["Mileage(Km)"]=cars_df["Mileage(Km)"].str.replace(r'[^\d.]', '', regex=True).astype(float)
cars_df["Tax Horsepower(CV)"]=cars_df["Tax Horsepower(CV)"].str.replace(r'[^\d.]', '', regex=True).astype(float)
cars_df
```

```python
import json
import uuid

cars_df.insert(0,"_id",[str(uuid.uuid4()) for _ in range(len(cars_df))])
with open('cars.json', 'r') as file:
    existing_data = json.load(file)


existing_records = {json.dumps(record, sort_keys=True) for record in existing_data}


new_data = cars_df.to_dict(orient='records')


unique_new_data = [
    record for record in new_data
    if json.dumps(record, sort_keys=True) not in existing_records
]


existing_data.extend(unique_new_data)


with open('cars.json', 'w') as file:
    json.dump(existing_data, file, indent=4)
```
