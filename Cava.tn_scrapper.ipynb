```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from cleantext import clean


cars_df=pd.DataFrame(columns=["Model","Price(DT)","Mileage(Km)","Year","Gearbox","Transmission","Tax Horsepower(CV)","Energy","Governorate","Description","Image","Original Post"])
```

```python
cars_urls=[]
size=100
for i in range(1,size):
    print(f'{(i/size)*100}%')
    url=f"https://www.cava.tn/category/voitures?page={i}"
    page=requests.get(url)
    scrapped=BeautifulSoup(page.text,"html.parser")
    articles=scrapped.find_all('app-product-item')
    for article in articles:  
        cars_urls.append("https://www.cava.tn"+article.find("a").get("href"))
```

```python
i=0
for url in cars_urls:
    i+=1
    if i%10==0:
        print(f'{round(i/len(cars_urls)*100,2)}%')
    page=requests.get(url)
    scrapped=BeautifulSoup(page.text,"html.parser")
    name=clean(scrapped.find("h1").text,no_emoji=True)
    price="0"
    if scrapped.find("app-product-details").find("div",class_="price font-p-s-b"):
        price=(scrapped.find("app-product-details").find("div",class_="price font-p-s-b").find("label").text.replace("TND","").replace("."," "))
    mileage="0"
    year=""
    gearbox=""
    tax_horsepower="0"
    energy=""
    governorate="None"
    if scrapped.find("app-product-details").find("div",class_="px-3") and scrapped.find("app-product-details").find("div",class_="px-3").find("label").find("a") :
        governorate=scrapped.find("app-product-details").find("div",class_="px-3").find("label").find("a").text
    description="None"
    if scrapped.find("app-product-details").find("div",class_="product-specify product-description p-3 col-12 col-md-12 cv-text-start mt-1") and scrapped.find("app-product-details").find("div",class_="product-specify product-description p-3 col-12 col-md-12 cv-text-start mt-1").find("p") :
        description=clean(scrapped.find("app-product-details").find("div",class_="product-specify product-description p-3 col-12 col-md-12 cv-text-start mt-1").find("p").text,no_emoji=True)
    image="None"
    if scrapped.find("app-product-details").find("div",class_="cv-gallery-img-container w-100 ng-star-inserted"):
        image=scrapped.find("app-product-details").find("div",class_="cv-gallery-img-container w-100 ng-star-inserted").find("img").get("src")
    link=url
    criterias=scrapped.find("app-product-details").find("div",class_="d-flex align-items-center product-specify-container").find_all("div") if scrapped.find("app-product-details").find("div",class_="d-flex align-items-center product-specify-container") else []
    if criterias:
        for criteria in criterias:
            criteria_name=criteria.find("span",class_="font-p-r").text
            if criteria_name=="Kilométrage":
                mileage=criteria.find("a").text+" km"
                
            elif criteria_name=="Boîte de vitesse":
                gearbox=criteria.find("a").text
               
            elif criteria_name=="Année":
                year=criteria.find("a").text
                
            elif criteria_name=="Puissance fiscale":
                tax_horsepower=criteria.find("a").text
            
            elif criteria_name=="Carburant":
                energy=criteria.find("a").text
        row={"Model":name,"Price(DT)":price,"Mileage(Km)":mileage,"Year":year,"Gearbox":gearbox,"Transmission":"None","Tax Horsepower(CV)":tax_horsepower,"Energy":energy,"Governorate":governorate,"Description":description,"Image":image,"Original Post":link}
        cars_df.loc[len(cars_df)]=row
                          
           
    

        
```

```python
cars_df=cars_df.drop_duplicates()
cars_df["Price(DT)"] = cars_df["Price(DT)"].str.replace(r'[^\d.]', '', regex=True).astype(float)
cars_df["Mileage(Km)"]=cars_df["Mileage(Km)"].str.replace(r'[^\d.]', '', regex=True).astype(float)
cars_df["Tax Horsepower(CV)"]=cars_df["Tax Horsepower(CV)"].str.replace(r'[^\d.]', '0', regex=True).astype(float)
cars_df
```

```python
import json
import uuid 

cars_df.insert(0,"_id",[str(uuid.uuid4()) for _ in range(len(cars_df))])
with open('cars.json', 'r') as file:
    existing_data = json.load(file)


existing_records = {json.dumps(record, sort_keys=True) for record in existing_data}


new_data = cars_df.to_dict(orient='records')


unique_new_data = [
    record for record in new_data
    if json.dumps(record, sort_keys=True) not in existing_records
]


existing_data.extend(unique_new_data)


with open('cars.json', 'w') as file:
    json.dump(existing_data, file, indent=4)
```
