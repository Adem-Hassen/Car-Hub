```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from cleantext import clean
```

```python
cars_urls=[]
size=500
for i in range(1,size):
    if i%10==0:
        print(f'{(i/size)*100}%')

    url=f"https://www.tayara.tn/ads/c/Véhicules/Voitures/?page={i}"
    page=requests.get(url)
    scrapped=BeautifulSoup(page.text,"html.parser")
    articles=scrapped.find_all('article')
    for article in articles:  
        cars_urls.append("https://www.tayara.tn"+article.find("a").get("href"))
```

```python
cars_df=pd.DataFrame(columns=["Model","Price(DT)","Mileage(Km)","Year","Gearbox","Transmission","Tax Horsepower(CV)","Energy","Governorate","Description","Image","Original Post"])
i=0
for url in cars_urls:
    i+=1
    if i%10==0:
        print(f'{round(i/len(cars_urls)*100,2)}%')
    page=requests.get(url)
    scrapped=BeautifulSoup(page.text,"html.parser")
    name=clean(scrapped.find("h1").text,no_emoji=True)
    if scrapped.find("data") and scrapped.find("data").get("value"):
        price=scrapped.find("data").get("value")
    else: 
        price="0"
    mileage="0"
    year=""
    gearbox=""
    tax_horsepower="0"
    energy=""
    if scrapped.find("div",class_="flex items-center space-x-2 mb-1"):
        governorate=scrapped.find("div",class_="flex items-center space-x-2 mb-1").find("span").text.split(",")[0]
    else:
        governorate="None"
    if scrapped.find("p",class_="text-sm text-start text-gray-700 font-arabic whitespace-pre-line line-clamp-3"):
        description=clean(scrapped.find("p",class_="text-sm text-start text-gray-700 font-arabic whitespace-pre-line line-clamp-3").text,no_emoji=True)
    else:
        description="None"
    if scrapped.find("img",class_="w-full h-full object-cover cursor-pointer"):      
        image=scrapped.find("img",class_="w-full h-full object-cover cursor-pointer").get("src")
    else:
        image="None"
    link=url

    criterias=scrapped.find("ul",class_="grid gap-3 grid-cols-12").find_all("li") if scrapped.find("ul",class_="grid gap-3 grid-cols-12") else []
    if criterias:
        for criteria in criterias:
            criteria_name=criteria.find("span",class_="text-gray-600/80 text-2xs md:text-xs lg:text-xs font-medium").text
            if  criteria_name in ["Couleur du véhicule","Etat du véhicule","Cylindrée","Marque","Modèle","Type de carrosserie"]:
                continue
            if criteria_name=="Kilométrage":
                if criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold"):
                    mileage=criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold").text 
                else:
                    mileage="0"
            elif criteria_name=="Boite":
                gearbox=criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold").text
            elif criteria_name=="Année":
                year=criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold").text
            elif criteria_name=="Puissance fiscale":
                if criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold"):
                
                    tax_horsepower=criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold").text if criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold") else "0"
            elif criteria_name=="Carburant":
                energy=criteria.find("span",class_="text-gray-700/80 text-xs md:text-sm lg:text-sm font-semibold").text

        row={"Model":name,"Price(DT)":price,"Mileage(Km)":mileage,"Year":year,"Gearbox":gearbox,"Transmission":"None","Tax Horsepower(CV)":tax_horsepower,"Energy":energy,"Governorate":governorate,"Description":description,"Image":image,"Original Post":link} 
    
        cars_df.loc[len(cars_df)]=row
        
         
    
```

```python
cars_df=cars_df.drop_duplicates()
cars_df["Price(DT)"] = cars_df["Price(DT)"].str.replace(r'[^\d.]', '', regex=True).astype(float)
cars_df["Mileage(Km)"]=cars_df["Mileage(Km)"].str.replace(r'[^\d]', '', regex=True).astype(float)
cars_df["Tax Horsepower(CV)"]=cars_df["Tax Horsepower(CV)"].str.replace(r'[^\d]', '', regex=True).astype(float)
```

```python
import json
import uuid

cars_df.insert(0,"_id",[str(uuid.uuid4()) for _ in range(len(cars_df))])
with open('cars.json', 'r') as file:
    existing_data = json.load(file)


existing_records = {json.dumps(record, sort_keys=True) for record in existing_data}


new_data = cars_df.to_dict(orient='records')


unique_new_data = [
    record for record in new_data
    if json.dumps(record, sort_keys=True) not in existing_records
]


existing_data.extend(unique_new_data)


with open('cars.json', 'w') as file:
    json.dump(existing_data, file, indent=4)
```
